```python

!pip install -q google-generativeai gradio

import google.generativeai as genai
import gradio as gr
import os

# خدي الـ API Key من environment (مش في الكود)
genai.configure(api_key=os.getenv("GEMINI_API_KEY"))

model = genai.GenerativeModel("gemini-1.5-flash")

chat = model.start_chat(history=[])

def chatbot(user_input, history):
    response = chat.send_message(user_input)
    history.append((user_input, response.text))
    return history, history

def user_message(user_input, history):
    history.append((user_input, ""))
    return "", history

with gr.Blocks() as demo:
    gr.Markdown("# 🤖 Smart AI Chatbot")

    chatbot_ui = gr.Chatbot()
    msg = gr.Textbox(placeholder="Type your message here...")

    state = gr.State([])

    msg.submit(user_message, [msg, state], [msg, chatbot_ui]).then(
        chatbot, [msg, state], [chatbot_ui, state]
    )

demo.launch()
```
